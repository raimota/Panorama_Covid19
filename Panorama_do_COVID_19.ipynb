# Panorama do COVID-19

A COVID-19 emergiu como uma preocupação global desde que o novo coronavírus, conhecido como SARS-CoV-2, foi identificado pela primeira vez em Wuhan, China, no final de 2019. O alerta inicial foi dado pela Organização Mundial da Saúde (OMS) em 31 de dezembro daquele ano, quando um conjunto de casos de "pneumonia viral" foi relatado.

A propagação desse vírus ocorre principalmente por meio de gotículas respiratórias expelidas durante tosses, espirros e até mesmo ao falar. Essas gotículas, quando inaladas por indivíduos saudáveis, podem resultar na infecção pelo vírus. A complexidade da COVID-19 se manifesta na ampla gama de sintomas que varia de pessoa para pessoa, desde casos assintomáticos até condições graves que requerem cuidados médicos intensivos.

<img alt="Colaboratory logo" width="80%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg">

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados neste conjunto são fornecidos pelo Our World in Data (OWID), uma iniciativa colaborativa que visa tornar os dados sobre questões globais, como saúde, desenvolvimento e meio ambiente, acessíveis e compreensíveis para todos. O OWID coleta dados de várias fontes confiáveis, como organizações de saúde nacionais e internacionais, governos, e agências de estatísticas. Eles são conhecidos por sua transparência e rigor na coleta e análise de dados, o que os torna uma fonte confiável para pesquisadores, jornalistas e tomadores de decisão em todo o mundo.

Este conjunto de dados abrange uma ampla gama de métricas relacionadas à pandemia de COVID-19, incluindo casos confirmados, mortes, testes, vacinações, medidas governamentais de resposta, características demográficas e socioeconômicas, e muito mais. Ele oferece uma visão abrangente da propagação da doença e de suas consequências em diferentes países e regiões ao redor do mundo.

Além dos números básicos de casos e mortes, o dataset também inclui informações sobre a capacidade de resposta dos sistemas de saúde, como o número de leitos hospitalares disponíveis, e fatores de risco relacionados à saúde, como prevalência de doenças crônicas e hábitos de tabagismo.

Os dados são atualizados regularmente e são apresentados de forma organizada e acessível, facilitando análises comparativas entre países e ao longo do tempo.

Analisar uma doença nova, como foi o caso da COVID-19, apresenta uma série de desafios significativos devido à natureza desconhecida e em rápida evolução da doença. Algumas das principais dificuldades incluem:

1.   Falta de Dados Consistentes e Confiança: No início de uma pandemia, os dados disponíveis podem ser limitados, incompletos ou inconsistentes. Isso pode dificultar a compreensão da extensão da propagação da doença e a formulação de estratégias eficazes de prevenção e controle.

2.   Mudanças nas Diretrizes e Protocolos: À medida que os cientistas aprendem mais sobre a doença e novas evidências emergem, as diretrizes e protocolos para diagnóstico, tratamento e relatórios de dados podem mudar. Isso pode levar a discrepâncias nos números relatados ao longo do tempo e entre diferentes fontes.

3. Variação na Testagem e Notificação: A capacidade de testagem e a abrangência da notificação de casos podem variar amplamente entre diferentes regiões e países. Isso pode resultar em subnotificação de casos reais e distorcer a imagem verdadeira da propagação da doença.

4. Interesses Políticos e Econômicos: Em alguns casos, os governos podem ter incentivos para minimizar ou exagerar a gravidade da situação por motivos políticos, econômicos ou de imagem pública. Isso pode levar a conflitos entre os dados oficiais divulgados pelos governos federal e estaduais e os dados independentes coletados por organizações não governamentais ou pela comunidade científica.

5. Desafios na Padronização e Harmonização dos Dados: Mesmo quando os dados são relatados de forma consistente pelos governos, pode haver desafios na padronização e harmonização dos dados entre diferentes jurisdições e países. Isso pode dificultar a comparação direta e a análise agregada dos dados.



No contexto da COVID-19, esses desafios foram amplificados pela velocidade com que a doença se espalhou globalmente e pela necessidade urgente de tomar decisões informadas para conter sua propagação. Portanto, a análise de uma doença nova como a COVID-19 requer uma abordagem cuidadosa e crítica, levando em consideração as limitações e incertezas associadas aos dados disponíveis.<br><br>

1. **Natureza Volátil dos Dados Emergenciais:** Durante uma crise como a pandemia de COVID-19, a coleta e o registro de dados podem ser desafiadores devido à rápida evolução da situação. Isso pode levar a lacunas na documentação de casos, internações em UTI e mortalidade excessiva.

Isso pode resultar em lacunas na documentação de internações hospitalares e na atualização de estatísticas relacionadas.

4. **Desafios Logísticos:** Restrições logísticas, como acesso limitado a áreas afetadas, interrupções nos sistemas de transporte e falta de pessoal qualificado, podem prejudicar a coleta e o relato oportuno de dados.

5. **Mudanças nas Prioridades de Coleta de Dados:** Durante a pandemia, as autoridades de saúde podem ter revisado suas prioridades de coleta de dados, concentrando-se em métricas mais imediatas, como novos casos e mortes, em detrimento de variáveis menos urgentes, como hospitalizações e mortalidade excessiva.

### Gráfico: Dispersão do percentual de vacinados em relação ao percentual de óbitos.

```
# @title
# Calcular as proporções
df_br['percentual_mortes'] = df_br['total_deaths'] / df_br['total_cases'].replace(0, 1)
df_br['percentual_vacinados'] = df_br['people_vaccinated'] / df_br['population'].replace(0, 1)

# Filtrar os dados a partir da primeira data de vacinação no Brasil
df_filtrado = df_br.loc[df_br['date'] >= '2021-01-17']

# Definir os limites das faixas de percentual de vacinação
faixas_vacinacao = [(0, 0.25), (0.26, 0.50), (0.51, 0.75), (0.76, 1)]

# Criar categorias com base nos limites definidos
df_filtrado['faixa_vacinacao'] = pd.cut(df_filtrado['percentual_vacinados'], bins=[faixa[0] for faixa in faixas_vacinacao] + [1], labels=["0-25%", "26-50%", "51-75%", "76-100%"])

# Configurar o tamanho da figura
fig = plt.gcf()
fig.set_size_inches(15, 8)

# Criar o gráfico de dispersão
ax = sns.scatterplot(data=df_filtrado, x="percentual_vacinados", y="percentual_mortes", hue="faixa_vacinacao", palette="magma", s=100)
ax.set(xlabel='Percentual de vacinados', ylabel='Percentual de mortes')
ax.set_title("Percentual de vacinados x Percentual de mortes")
ax.grid(False)

# Definir os limites e rótulos dos ticks no eixo x
intervalos = np.arange(0, 1.1, 0.25)
ax.set_xticks(intervalos)
ax.set_xticklabels(["0%", "25%", "50%", "75%", "100%"])


plt.show()
```
